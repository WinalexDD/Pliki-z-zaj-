```python
def get_all_tag_names(document):
    """Returns list with all tag names"""
    elements = document.getElementsByTagName("*")
    return set(map(lambda x: x.tagName, elements))
```

```python
get_all_tag_names(document)
```

```python
#gorsza wersja
tags=set()
def get_all_tag_names2(element):
    if element.nodeType== Node.ELEMENT_NODE:
        tags.add(element.tagName)
    for e in element.childNodes:
        get_all_tag_names2(e)
        
    return tags
```

```python
get_all_tag_names2(document)
```

```python
#lekka naprawa gorszej wersja
def get_all_tag_names3(document):
    def _get_all_tag_names3(element):
        if element.nodeType== Node.ELEMENT_NODE:
            tags.add(element.tagName)
        for e in element.childNodes:
            get_all_tag_names3(e)
    tags=set()
    return _get_all_tag_names3(document)
```

```python
get_all_tag_names3(document)
```

```python
#jeszcze inna
def get_all_tag_names4(element):
    tags=set()   
    if element.nodeType== Node.ELEMENT_NODE:
        tags.add(element.tagName)
        
    tags = tags.union(*[get_all_tag_names4(e) for e in element.childNodes])
    return tags 
```

```python
get_all_tag_names4(document)
```

```python
def get_protein_name(xml_file):
    document=parse(xml_file)
    
    entry=document.getElementsByTagName("entry")[0]
    protein=entry.getElementsByTagName("protein")[0]
    recommended_name=protein.getElementsByTagName("recommendedName")[0]
    fullName=recommended_name.getElementsByTagName("fullName")[0]
    
    return fullName
```

```python
nazwa=[]
n=0

for i in lista:
    response = requests.get('https://www.uniprot.org/uniprotkb/'+i+'.xml')
    if response.status_code == 200:
        dom = parseString(response.text)
        full_names = dom.getElementsByTagName("fullName")

        if full_names:
            full_name = full_names[0].firstChild.wholeText
            nazwa.append(full_name)
        else:
            print(f"No fullName found for {i}")
            nazwa.append(None)
    else:
        print(f"Error fetching data for {i}: HTTP {response.status_code}")
        nazwa.append(None)
    n+=1
```

```python
def fetch_uniprot_data(protein_id):
    url=f'https://www.uniprot.org/uniprot/{protein_id}.xml'
    response=requests.get(url)
    
    if response.status_code==200:
        return response.text
    else:
        return "Nie udalo sie pobrac danych"
    
def get_protein_name_from_text(protein_data):
    document=parseString(protein_data)
    return document.getElementsByTagName("fullName")[0].firstChild.nodeValue

df = pd.read_csv("proteins.csv")

df["Identifier"]=df["Protein"].apply(lambda x: x.split("|")[1])
protein_names=[]
for protein_id in df["Identifier"]:
    protein_data=fetch_uniprot_data(protein_id)
    protein_name=get_protein_name_from_text(protein_data)
    protein_names.append(protein_name)
    
print(protein_names)
```

```python
from xml.sax import parse, SAXException
from xml.sax.handler import ContentHandler
```

```python
class DictHandler(ContentHandler):
    """Parses XML to create a dictionary of the following format:
    {'name': name of the tag
    'attributes': dict with attributes
    'children': list of children
    'value': value inside tag}"""
        
    def __init__(self):
        super().__init__()
        self.element_stack = []
        
    def startElement(self, name, attrs):
        "Run when opening tag is found"
        self.element_stack.append({
            'name': name,
            'attributes': dict(attrs),
            'children': [],
            'value': ''})
    
    def endElement(self, name):
        if len(self.element_stack) > 1:
            child = self.element_stack.pop()
            self.current_element['children'].append(child)
    
    def characters(self, content):
        "Run when string is found"
        if self.element_stack:
            self.current_element['value'] += content
        
    @property
    def current_element(self):
        return self.element_stack[-1] if self.element_stack else None
```

```python
handler=DictHandler()
parse('./P01024.xml', handler)
root=handler.current_element
```
